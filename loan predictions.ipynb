```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
s=pd.read_csv('t1train.csv')
s.drop(['Loan_ID'],inplace=True,axis=1)
```

```python
s
```

```python
s.isnull().sum()
```

```python
s['Gender'].fillna(s['Gender'].mode()[0],inplace=True)
```

```python
s['Credit_History'].fillna(s['Credit_History'].mode()[0],inplace=True)
```

```python
s['Dependents'].fillna(s['Dependents'].mode()[0],inplace=True)
```

```python
s['Self_Employed'].fillna(s['Self_Employed'].mode()[0],inplace=True)
```

```python
s['LoanAmount'].fillna(
    
    value=128.0,
    axis=0,
    inplace=True,
    )
```

```python
s
```

```python
s['Loan_Amount_Term'].fillna(s['Loan_Amount_Term'].mode()[0],inplace=True)
```

```python
s
```

```python
s['Married']=pd.get_dummies(s['Married'],drop_first=True)
s['Education']=pd.get_dummies(s['Education'],drop_first=True)
s['Self_Employed']=pd.get_dummies(s['Self_Employed'],drop_first=True)
s['Property_Area']=pd.get_dummies(s['Property_Area'],drop_first=True)
s['Credit_History']=pd.get_dummies(s['Credit_History'],drop_first=True)
s['Dependents']=pd.get_dummies(s['Dependents'],drop_first=True)
s['Gender']=pd.get_dummies(s['Gender'],drop_first=True)
```

```python
s
```

```python
s.isnull().sum()
```

```python
s['Total_income']=s['ApplicantIncome']+s['CoapplicantIncome']
```

```python
s['EMI']=(s['LoanAmount']*0.09*(1.09**s['Loan_Amount_Term']))/(1.09**(s['Loan_Amount_Term']-1))
```

```python
s.drop(['ApplicantIncome','CoapplicantIncome'],axis=1,inplace=True)
```
